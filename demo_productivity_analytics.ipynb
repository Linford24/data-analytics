````python
class ProductivityAnalysis:
    def __init__(self):
        self.data = pd.read_csv("gen_data.csv")
        self.mean_metrics = ["Task Completion Rate %",  "Engagement Level"]
        self.count_metrics = ["Resource Utilization"]
        
        
    def data_columns(self):
        """
        Prints the data columns in the csv file.
        
        """
        if self.data is not None:   
            return self.data.columns
        else:
            return "The data file is empty."
        
    def bar_visualization(self, result_metric, employee, metric):
        """
         Creates a bar visualization for analysis of employee productivity metrics
        
         Args:
           - employee:
               the employee(s) to be analysed.
           
           - metric:
               the metric being analysed relative to the employee(s)
           
           - result_metric: 
                 the type of arithmetic to to scale the metric by. Could be count() or sum() or mean()
        
        """
        
        # Create the bar visualization
        fig, ax = plt.subplots(figsize=(10, 0.5))

        # Plot the background bar (full length = 100%)
        ax.barh(y=0, width=100, height=0.5, color="lightgray", alpha=0.3)

        # Plot the filled portion (mean metric)
        ax.barh(y=0, width=result_metric, height=0.5, color="blue", alpha=0.6)

        #Add "Mean Metric" label at the top of the bar
        ax.text(0, 0.6, f"{employee}'s {metric}", ha="left", va="center", color="black", fontsize=12, fontweight="bold")

        # Add the mean value at the end of the bar
        ax.text(result_metric + 60, 0, f"{result_metric:.2f}%", ha="left", va="center", color="black", fontsize=12, fontweight="bold")
                               
        # Customize the plot
        ax.set_xticks([])
        ax.set_yticks([]) 
        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_visible(False)
        

    def employee_bar_analysis(self, employee, metric):
        
        """
         Creates a bar visualization for analysis of employee productivity metrics
        
         Args:
           - employee:
               the employee(s) to be analysed.
           
           - metric:
               the metric being analysed relative to the employee(s)
        
        """
        
        if metric in self.mean_metrics:
            #Calculate mean metric for the employee 
            employee_metric_mean = self.data[self.data["Employee"] == employee][metric].mean()
            return self.bar_visualization(employee_metric_mean, employee, metric)
        
        elif metric in self.count_metrics:
            #Calculate count metric for the employee 
            employee_metric_count = self.data.groupby(self.data["Employee"])[metric].count()
            total_employees_metric_count = self.data[metric].count()
            count_metric = (employee_metric_count[employee] / total_employees_metric_count) * 1000
            return self.bar_visualization(count_metric, employee, metric)
        
        else:
            #Calculate sum metric for the employee 
            employee_metric_sum = self.data.groupby(self.data["Employee"])[metric].sum()
            total_employees_metric_sum = self.data[metric].sum()
            sum_metric = (employee_metric_sum[employee] / total_employees_metric_sum) * 1000
            return self.bar_visualization(sum_metric, employee, metric)
    
    def general_trend_analysis(self, groupby_metrics, metric, order_by, employee=None):
        """
         Perform general trend analysis on all employees.

         Parameters:
          - groupby_metrics: Metrics to group by.
          - metric: The metric to analyze.
          - order_by: The order in which to sort the results(either days or months)
          - employee: Optional. Specific employee or a list of employees to analyze. Defaults to None.
        """
        
        return self.trend_visualization(employee, groupby_metrics, metric,order_by)
    
    def employee_trend_analysis(self, employee, groupby_metrics, metric, order_by):
        """
         Perform trend analysis on selected employee(s) data.

         Parameters:
          - groupby_metrics: Metrics to group by.
          - metric: The metric to analyze.
          - order_by: The order in which to sort the results(either days or months)
          - employee: Optional. Specific employee or a list of employees to analyze. Defaults to None.
        """
        return self.trend_visualization(employee, groupby_metrics, metric,order_by)

    
       
    
    def trend_visualization(self, employee, groupby_metrics, metric, order_by):
        
        """
         Perform trend analysis on selected employee(s) data.

         Parameters:
          - groupby_metrics: Metrics to group by.
          - metric: The metric to analyze.
          - order_by: The order in which to sort the results(either days or months)
          - employee: Optional. Specific employee or a list of employees to analyze. Defaults to None.
        """
        
        if order_by == "months":
            order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        elif order_by == "days":
            order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        else:
            order = None
            
        if metric in self.mean_metrics:
            #Calculate mean metric for the employee 
            data_agg = self.data.groupby(groupby_metrics, as_index=False)[metric].mean()
            
        
        elif metric in self.count_metrics:
            #Calculate count metric for the employee 
            data_agg = self.data.groupby(groupby_metrics, as_index=False)[metric].count()
            
        
        else:
            #Calculate sum metric for the employee 
            data_agg = self.data.groupby(groupby_metrics, as_index=False)[metric].sum()
        
        data_agg[order_by] = pd.Categorical(data_agg[order_by], categories=order, ordered=True)
        
        plt.figure(figsize=(20, 6))
        
        if employee is not None:
            if isinstance(employee, list): 
                data_filtered = data_agg[data_agg["Employee"].isin(employee)]
                sns.lineplot(data=data_filtered, x=order_by, y=metric, hue="Employee", marker="o")
            
            else:
                data_filtered = data_agg[data_agg["Employee"] == employee]
                sns.lineplot(data=data_filtered, x=order_by, y=metric, hue="Employee", marker="o")
        else:
            sns.lineplot(data=data_agg, x=order_by, y=metric, hue="Employee", marker="o")
            employee = "all"

         # Add title and labels
        plt.title(f"Total {metric} by {order_by} for {employee}")
        plt.xlabel(f"{order_by} of the Week")
        plt.ylabel(f"Total {metric}")

        # Rotate x-axis labels for better readability
        plt.xticks(rotation=45)

        # Show the legend
        plt.legend(title="Employee", bbox_to_anchor=(1.05, 1), loc="upper left")
        # Display the plot
        plt.tight_layout()
        
        
    
            
   
        
````

````python
import openai
import json

def preprocess_conversation_data(full_conversation):
    """Convert structured conversation data into text segments"""
    segments = []
    for message in full_conversation.get('messages', []):
        parts = []
        
        if 'chatMessage' in message:
            cm = message['chatMessage']
            sender = cm.get('sender', {})
            parts.append(f"Chat from {sender.get('name', 'Unknown')} ({sender.get('departmentName', 'Unknown')}): {cm.get('content', '')}")
            
            if 'reply' in cm:
                reply = cm['reply']
                parts.append(f"[Reply by {reply.get('replier', 'Unknown')} ({reply.get('departmentName', 'Unknown')}) at {reply.get('replyTime', 'unknown time')}]")
        
        elif 'voiceCall' in message:
            vc = message['voiceCall']
            parts.append(format_call_data(vc, "Voice call"))
        
        elif 'videoCall' in message:
            vc = message['videoCall']
            parts.append(format_call_data(vc, "Video call"))
        
        if parts:
            segments.append(' '.join(parts))
    
    return segments

def format_call_data(call_data, call_type):
    """Format call information into natural language"""
    caller = call_data.get('caller', {})
    recipient = call_data.get('recipient', {})
    return (
        f"{call_type} from {caller.get('name', 'Unknown')} ({caller.get('departmentName', 'Unknown')}) "
        f"to {recipient.get('name', 'Unknown')} ({recipient.get('departmentName', 'Unknown')}) "
        f"lasting {call_data.get('duration', 'unknown duration')}"
    )

def identify_json_structure(conversation_segment):
    """More sensitive categorization with examples"""
    prompt = f"""
    Analyze this workplace communication and categorize it:
    {conversation_segment}

    Categories (respond with ONLY these exact phrases):
    1. "Productivity analytics" - if discussing metrics like:
       - Efficiency, throughput, completion rates
       - Time tracking, focus periods
       - "We finished 15% more tickets"
    2. "Automation optimization" - if discussing:
       - Workflows, bots, RPA
       - "The approval process takes too long"
    3. "Business health" - if discussing:
       - CSAT, NPS, retention
       - "Our customer satisfaction dropped"
       - Department performance
    4. "None" - for all other cases

    Conversation contains:"""
    
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message['content'].strip()

def extract_structured_data(conversation_segment, category):
    """More flexible extraction with implied metrics"""
    if category == "None":
        return None

    templates = {
        "Productivity analytics": {
            "metrics": ["efficiency", "throughput", "focus_time", "completion_rate"],
            "template": {"productivityAnalytics": []}
        },
        "Automation optimization": {
            "metrics": ["process_time", "bottlenecks", "automation_potential"],
            "template": {"automationOptimization": []}
        },
        "Business health": {
            "metrics": ["csat", "nps", "retention", "response_time"],
            "template": {"businessHealth": []}
        }
    }

    prompt = f"""
    Extract {category} insights from:
    {conversation_segment}

    Possible metrics: {", ".join(templates[category]["metrics"])}

    Return JSON with:
    - metric names as keys
    - values as numbers/strings
    - context like departments/timeframes

    Example for CSAT drop:
    {{"csat": 0.82, "response_time": "increased", "department": "Customer Support"}}"""
    
   
    openai.api_key = OPENAI_API_KEY
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        response_format={ "type": "json_object" }  
    )

    try:
        extracted = json.loads(response.choices[0].message['content'])
        category_key = list(templates[category]["template"].keys())[0]
        templates[category]["template"][category_key].append(extracted)
        return templates[category]["template"]
    except json.JSONDecodeError:
        return None



def process_conversation(full_conversation):
    """Main processing pipeline"""
    segments = preprocess_conversation_data(full_conversation)
    
    results = {
        "productivityAnalytics": [],
        "automationOptimization": [],
        "businessHealth": []
    }
    
    for segment in segments:
        category = identify_json_structure(segment)
        if category == "None":
            continue
        
        extracted = extract_structured_data(segment, category)
        if not extracted:
            continue
        
        key = list(extracted.keys())[0]
        results[key].extend(extracted[key])
    
    return {k: v for k, v in results.items() if v}

def normalize_category_name(category: str) -> str:
    """Standardize category names for dictionary lookup"""
    category = category.strip().lower()
    if "productivity" in category:
        return "Productivity analytics"
    elif "automation" in category:
        return "Automation optimization"
    elif "business" in category:
        return "Business health"
    return "None"

def extract_structured_data(conversation_segment, category):
    """Handle JSON output without response_format parameter"""
    normalized_category = normalize_category_name(category)
    if normalized_category == "None":
        return None
    
    templates = {
        "Productivity analytics": {
            "metrics": ["workforceEfficiency", "taskCompletionRate"],
            "output_key": "productivityAnalytics"
        },
        "Automation optimization": {
            "metrics": ["automationCoverage", "processBottlenecks"],
            "output_key": "automationOptimization"
        },
        "Business health": {
            "metrics": ["customerSatisfaction", "employeeRetention"],
            "output_key": "businessHealth"
        }
    }
    
    config = templates.get(normalized_category)
    if not config:
        return None

    prompt = f"""Extract {normalized_category} metrics from this message:
      {conversation_segment}

     Available metrics: {", ".join(config["metrics"])}

     Return ONLY a valid JSON object with:
     1. At least one metric from the list
     2. Numeric values where possible
     3. Additional context like:
       - "department" (when mentioned)
       - "timeframe" (when mentioned)
       - "reason" (for changes)

       Example Output:
     ```json
       {{
          "{config["metrics"][0]}": 0.85,
          "department": "Customer Support",
          "trend": "decreasing"
        }}
      ```"""

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",  # Using standard GPT-4
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3  # More deterministic output
        )
        
        # Extract JSON from response
        raw_output = response.choices[0].message['content']
        json_str = raw_output.split("```json")[-1].split("```")[0].strip()
        extracted = json.loads(json_str)
        
        # Validate and structure output
        if not any(metric in config["metrics"] for metric in extracted.keys()):
            return None
            
        return {config["output_key"]: [extracted]}
        
    except (json.JSONDecodeError, KeyError, AttributeError) as e:
        print(f"Failed to parse LLM output: {e}")
        print(f"Raw output was: {raw_output}")
        return None



def normalize_data(raw_data):
    """Standardize extracted data formats"""
    normalized = {}
    
    for category, entries in raw_data.items():
        normalized[category] = []
        for entry in entries:
            normalized_entry = {}
            for metric, value in entry.items():
                if isinstance(value, (int, float)):
                    normalized_entry[metric] = value
                elif isinstance(value, str) and '%' in value:
                    normalized_entry[metric] = float(value.replace('%', '')) / 100
                else:
                    normalized_entry[metric] = str(value)
            normalized[category].append(normalized_entry)
    
    return normalized
````
